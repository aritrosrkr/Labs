```python
import heapq
import numpy as np

# n, m = map(int, input("Shape of Maze: ").split())
# a, b = map(int, input("Start Node: ").split())
# c, d = map(int, input("Goal Node: ").split())

# graph = []
# for i in range(n):
#     graph.append(list(input()))
# print(graph)

with open('input.txt', 'r') as f:
    lines = f.read().splitlines()

n, m = map(int, lines[0].split())
a, b = map(int, lines[1].split())
c, d = map(int, lines[2].split())
graph = [list(line) for line in lines[3:3+n]]

print(n, m)
print(a, b)
print(c, d)

def manhattan(n1, n2):
    return abs(n1[0] - n2[0]) + abs(n1[1] - n2[1])

def A_Star(graph, s, g):
    if graph[s[0]][s[1]] == '#' or graph[g[0]][g[1]] == '#':
        print(-1)
        return -1
    
    Q = []
    niggers = [(-1, 0, 'U'), (1, 0, 'D'), (0, -1, 'L'), (0, 1)]
    parent = {}
    visited = {}
    distance = {}
    moves = {'U': [-1,0], 'D':[1,0], 'L': [0,-1], 'R':[0,1]}
    
    heuristic = {}
    for i in range(n):
        for j in range(m):
            if graph[i][j] == '0':
                heuristic[(i, j)] = manhattan(g, (i, j))
                visited[(i, j)] = 0
                distance[(i, j)] = float('inf')
                parent[(i, j)] = None
    
    distance[s] = 0

    heapq.heappush(Q, (heuristic[s], 0, [s[0], s[1]]))
    while Q:
        h, c, u = heapq.heappop(Q)
        
        if u == g:
            break
        
        if graph[u[0]][u[1]] != '#' and visited[(u[0], u[1])] == 0:
            U = list(map(lambda x, y: x + y, u, moves['U']))
            D = list(map(lambda x, y: x + y, u, moves['D']))
            L = list(map(lambda x, y: x + y, u, moves['L']))
            R = list(map(lambda x, y: x + y, u, moves['R']))

            for dx, dy, dir in niggers:
                x = u[0] + dx
                y = u[1] + dy
                if 0 <= x < n and 0 <= y < m and graph[(x,y) == '0']:
                    if visited[(x,y)] == 0 and c+1 < distance[(x,y)]:
                        distance[(x,y)] = c+1
                        parent[(x,y)] = U
                        heapq.heappush(Q, (heuristic[(x,y)] + c+1, c+1, (x,y)))
                
            # if 0 <= U[0] < n and 0 <= U[1] < m and graph[U[0]][U[1]] == '0' and visited[(U[0], U[1])] == 0:
            #     if c+1 < distance[(U[0], U[1])]:
            #         distance[(U[0], U[1])] = c+1
            #         heapq.heappush(Q, (heuristic[(U[0], U[1])] + c+1, c+1, U))
            #         parent[(U[0], U[1])] = u
                    
            # if 0 <= D[0] < n and 0 <= D[1] < m and graph[D[0]][D[1]] == '0' and visited[(D[0], D[1])] == 0:
            #     if c+1 < distance[(D[0], D[1])]:
            #         distance[(D[0], D[1])] = c+1
            #         heapq.heappush(Q, (heuristic[(D[0], D[1])] + c+1, c+1, D))
            #         parent[(D[0], D[1])] = u
                    
            # if 0 <= L[0] < n and 0 <= L[1] < m and graph[L[0]][L[1]] == '0' and visited[(L[0], L[1])] == 0:
            #     if c+1 < distance[(L[0], L[1])]:
            #         distance[(L[0], L[1])] = c+1
            #         heapq.heappush(Q, (heuristic[(L[0], L[1])] + c+1, c+1, L))
            #         parent[(L[0], L[1])] = u
                    
            # if 0 <= R[0] < n and 0 <= R[1] < m and graph[R[0]][R[1]] == '0' and visited[(R[0], R[1])] == 0:
            #     if c+1 < distance[(R[0], R[1])]:
            #         distance[(R[0], R[1])] = c+1
            #         heapq.heappush(Q, (heuristic[(R[0], R[1])] + c+1, c+1, R))
            #         parent[(R[0], R[1])] = u

            visited[(u[0], u[1])] = 1
    return parent

path = A_Star(graph, (a,b), (c,d))

if path != -1:
    current = (c, d)
    order = []
    while current != None and path[(current[0], current[1])] is not None:
        prev = path[(current[0], current[1])]
        if prev is None:
            break

        dx = current[0] - prev[0]
        dy = current[1] - prev[1]
        
        if dx == -1 and dy == 0:
            direction = 'U'
        elif dx == 1 and dy == 0:
            direction = 'D'
        elif dx == 0 and dy == -1:
            direction = 'L'
        elif dx == 0 and dy == 1:
            direction = 'R'
            
        order.append(direction)
        current = prev

    order = order[::-1]
    print(len(order))
    print(''.join(order))
```
